```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
 
import sys
sys.path.append('/Users/dcard/Documents/Git/ipywidgets/')

from ipywidgets import StaticInteract, RangeWidget, RadioWidget

def plot(alpha=0.05, power=0.7, n=1):
    R = np.arange(0,1.05,0.05)
    beta = 1-power
    PPV = R*(1-beta**n)/(R + 1 - (1-alpha)**n - R*beta**n)
    fig, ax = plt.subplots()
    ax.plot(R,PPV)
    ax.set_ylim(0,1)
    ax.set_xlabel('R')
    ax.set_ylabel('PPV')
    return fig
    
StaticInteract(plot,
               alpha=RangeWidget(0, 0.1, 0.05),
               power=RangeWidget(0.5,1,0.1),
               #n=RangeWidget(1,3,1)
               )
```
